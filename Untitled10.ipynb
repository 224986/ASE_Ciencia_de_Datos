```python
# import statements
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
# create blobs
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.6, random_state=50)
# create np array for data points
points = data[0]
# create scatter plot
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
```

```python
# import KMeans
from sklearn.cluster import KMeans
```

```python
# create kmeans object
kmeans = KMeans(n_clusters=4)
# fit kmeans object to data
kmeans.fit(points)
# print location of clusters learned by kmeans object
print(kmeans.cluster_centers_)
# save new clusters for chart
y_km = kmeans.fit_predict(points)
```

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(points, method='ward'))
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
# save clusters for chart
y_hc = hc.fit_predict(points)
```

La ubicación y la distancia de cada grupo se trazan en el dendrograma. El dendograma se puede usar para identificar grupos para cualquier número. Los comienzos del primer grupo (azul), segundo grupo (rojo) y tercer grupo se ven claramente en el dendrograma de arriba (verde). Solo los primeros tres grupos están codificados por colores arriba, pero si pasa al lado rojo del dendrograma, también puede ver dónde comenzó el cuarto grupo. El dendrograma continúa hasta que cada punto tiene su propio grupo distinto en ese momento.

```python
plt.scatter(points[y_hc ==0,0], points[y_hc == 0,1], s=100, c='red')
plt.scatter(points[y_hc==1,0], points[y_hc == 1,1], s=100, c='black')
plt.scatter(points[y_hc ==2,0], points[y_hc == 2,1], s=100, c='blue')
plt.scatter(points[y_hc ==3,0], points[y_hc == 3,1], s=100, c='cyan')
```

En este caso, los resultados del agrupamiento jerárquico y k-medias fueron bastante comparables. Sin embargo, este no es siempre el caso. El agrupamiento jerárquico aglomerativo tiene la ventaja de que, en general, produce resultados más precisos. La desventaja es que, en comparación con k-means, el agrupamiento jerárquico requiere más tiempo y recursos para lograrlo.
